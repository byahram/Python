```python
import pandas as pd
import numpy as np
```

```python
usage = pd.read_csv("./data/working/tpss_bcycl_od_statnhm_202406_통합_w_자치구.csv")
```

```python
usage.head()
```

```python
usage = usage.drop(columns=["전체_이용_분", "전체_이용_거리"], errors="ignore")
```

```python
usage = usage.dropna(subset=["시작_대여소_자치구", "종료_대여소_자치구"])
```

```python
usage.shape
```

```python
junggu_usage = usage[
    (usage["시작_대여소_자치구"] == "중구") | (usage["종료_대여소_자치구"] == "중구")
]
```

```python
junggu_usage.head()
```

```python
junggu_usage.shape
```

```python
df = junggu_usage.copy()
df.head()
```

```python
df.shape
```

```python
df["기준_날짜"] = pd.to_datetime(df["기준_날짜"], format="%Y%m%d")
df["주간"] = np.where(df["기준_날짜"].dt.dayofweek < 5, "평일", "주말")
```

```python
df.head()
```

```python
df["시간대"] = df["기준_시간대"] // 100
```

```python
df.shape
```

```python
df.head()
```

```python
same_rental = (
    df[(df["집계_기준"] == "출발시간") & (df["시작_대여소_자치구"] == "중구") & (df["시작_대여소_ID"] == df["종료_대여소_ID"])]
     .groupby(["시작_대여소_ID","시작_대여소명","주간","시간대"], as_index=False)["전체_건수"]
     .sum()
    .rename(columns = {
        "시작_대여소_ID" : "대여소_ID",
        "시작_대여소명" : "대여소명",
        "전체_건수" : "동일대여소대여반납"
    })
)
```

```python
same_rental.head(10)
```

```python
same_district = (
    df[(df["집계_기준"] == "출발시간") & (df["시작_대여소_자치구"] == "중구") & (df["시작_대여소_ID"] != df["종료_대여소_ID"]) & 
                (df["종료_대여소_자치구"] == "중구")]
     .groupby(["시작_대여소_ID","시작_대여소명","주간","시간대"], as_index=False)["전체_건수"]
     .sum()
    .rename(columns = {
        "시작_대여소_ID" : "대여소_ID",
        "시작_대여소명" : "대여소명",
        "전체_건수" : "중구내대여반납"
    })
)
```

```python
same_district.head(10)
```

```python
outflow_df = (
    df[(df["집계_기준"] == "출발시간") & (df["시작_대여소_자치구"] == "중구") & (df["종료_대여소_자치구"] != "중구")]
     .groupby(["시작_대여소_ID","시작_대여소명","주간","시간대"], as_index=False)["전체_건수"]
     .sum()
    .rename(columns = {
        "시작_대여소_ID" : "대여소_ID",
        "시작_대여소명" : "대여소명",
        "전체_건수" : "유출량"
    })
)
```

```python
outflow_df.head()
```

```python
inflow_df = (
    df[(df["집계_기준"] == "출발시간") & (df["종료_대여소_자치구"] == "중구") & (df["시작_대여소_자치구"] != "중구")]
     .groupby(["시작_대여소_ID","시작_대여소명","주간","시간대"], as_index=False)["전체_건수"]
     .sum()
    .rename(columns = {
        "시작_대여소_ID" : "대여소_ID",
        "시작_대여소명" : "대여소명",
        "전체_건수" : "유입량"
    })
)
```

```python
inflow_df.head()
```

```python
junggu_grouped = pd.merge(
    same_rental, same_district,
    on=["대여소_ID","대여소명","주간","시간대"],
    how="outer"
).fillna(0)
```

```python
junggu_grouped = pd.merge(
    junggu_grouped, outflow_df,
    on=["대여소_ID","대여소명","주간","시간대"],
    how="outer"
).fillna(0)
```

```python
junggu_grouped = pd.merge(
    junggu_grouped, inflow_df,
    on=["대여소_ID","대여소명","주간","시간대"],
    how="outer"
).fillna(0)
```

```python
junggu_grouped
```

```python
junggu_grouped["동일대여소대여반납"] = junggu_grouped["동일대여소대여반납"].astype(int)
junggu_grouped["중구내대여반납"] = junggu_grouped["중구내대여반납"].astype(int)
junggu_grouped["유입량"] = junggu_grouped["유입량"].astype(int)
junggu_grouped["유출량"] = junggu_grouped["유출량"].astype(int)
```

```python
junggu_grouped
```

```python
junggu_grouped = junggu_grouped.sort_values(["대여소_ID", "주간", "시간대"]).reset_index(drop=True)
junggu_grouped.head()
```

```python
junggu_grouped
```

```python
junggu_grouped = junggu_grouped.drop(["대여소명"], axis = 1)
```
