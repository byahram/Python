## 각 컬럼 소개
1. 자치구 - 자치구
2. 시간_구분 - 이용량을 기준으로 시간을 나눔
    - 활성시간 : 이용량이 제일 많은 시간때(8, 16, 17, 18, 19, 20)
    - 비활성시간 : 이용량이 제일 적은 시간떄( 1, 2, 3, 4, 5, 23)
    - 중간시간 : 이용량이 중간인 시간때(0, 6, 7, 9, 10, 11, 12, 13, 14, 15, 21, 22)
3. 대여_건수 : 따릉이의 대여건수 (개)
4. 반납_건수 : 따릉이의 반납건수 (개)
5. |대여-반납| : 대여건수와 반납 건수의 차이(각 자치구마다의 대여건수와 반납건수의 차이) (개)
6. 빈대여소의비율 : 각 시간때의 자치구마다의 빈대여소의 수 / 자치구의 전체 대여소의 수를 구한 뒤 시간_구분의 시간때의 평균 (%)
7. std : 빈 대여소의 시간으로 묶은 값의 표준편차
8. 빈대여소시간 : 각 자치구마다 빈 대여소가 몇시간 있는지 (시간)
9. 유출량 : 자치구에서 다른 자치구로 유출된 따릉이의 갯수 (개)
10. 유입량 : 자치구에서 다른 자치구로 부터 유입된 따릉이의 갯수 (개)
11. 회전율 : (대여건수 + 반납건수) / 각 자치구의 대여소의 수(얼마나 따릉이가 활발하게 사용되는지)
12. 최대 이용량 : 24시간중 가장 따릉이가 많이 이용된 갯수 (개)
13. 최소 이용량 : 24시간중 가장 따릉이가 적게 이용된 갯수 (개)
14. 최대최소_편차 : 최대이용량 - 최소이용량 (개)
15. 대여소의 개수 : 각 자치구마다의 대여소의 갯수 (개)
16. 면적 : 자치구의 면적 (km²)
17. 밀도 : 자치구의 대여소의 갯수 / 자치구의 면적 (개/km²)
18. 이동량 : |유입량 - 유출량|, 다른 자치구로부터 따릉이가 얼마나 이동했는지 (개)

## 입력변수
- 입력변수들은 따릉이 도메인 지식을 활용해서 변수 생성

- 각 자치구별 따릉이 수요 불균형을 나타내는 변수
  1. |대여-반납| : 각 자치구 별로 대여의 갯수와 반납의 갯수의 차이를 통해서 수요의 순흐름 규모 파악
     - 값이 크면 자전거 재고가 한쪽으로 쏠리기 쉬움
    
  2. 빈대여소의비율, 빈대여소시간 : 관측시간대의 빈대여소의 비율, 빈대여소가 얼마나 길게 이어졌는지의 시간을 파악
     - 불균형이 체감되는 빈도, 빈대여소의 비율이 높을 수록 시간이 클 수록 따릉이 불균형의 빈도를 확인할 수 있음
    
  3. 이동량 : 자치구간의 순외부흐름 파악
     - 이동량이 크다는 것은 자치구내의 유입이 많거나 유출이 많다는 것을 알 수 있음
    
  4. 최대최소_편차 : 시간대별 수요의 진폭을 파악
     - 피크타임과 비피크타임때의 격차가 클 수록 시간적인 면으로 불균형하다는 것을 확인 할 수 있음
    
- 각 자치구별 따릉이 사용 특징을 파악하기 위한 변수(직접적으로 불균형이 있다고 판단할 수 없는 변수)
  1. 회전율 : 단위 대여소 대비 처리된 이용량을 파악
     - 회전율이 높으면 따릉이 이용이 활발함
    
  2. 밀도 : 공간적인면에서 각 자치구별 대여소의 집중도를 파악

## 군집화 진행
1. 모든 변수들이 단위가 다르기 때문에 정규화를 진행해서 단위를 통일해줌
2. 엘보우기법으로 몇개의 군집이 최적인지 찾음
3. 엘보우 기법으로 찾은 k값을 기반으로 K-means 모델 생성
4. K-means로 군집화 진행

```python
x = df[select]
```

```python
#정규화
x_scaled = StandardScaler().fit_transform(x)
```

```python
# k를 2부터 10까지 해서 최적의 k 값 찾기
inertia = []

for k in range(2,10) :
    km = KMeans(n_clusters = k, random_state = 26)
    km.fit(x_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 10), inertia)
plt.xlabel("k")
plt.ylabel("inertia")
plt.show()
```

```python
# k를 3으로 넣고 k-means 모델생성
model = KMeans(n_clusters = 3, random_state = 26,  n_init=10)
```

```python
# 정규환된 입력변수를 이용해서 모델 학습
model.fit(x_scaled)
```

```python
# 0,1,2로 라벨링이되어있음
np.unique(model.labels_)
```

```python
group = df.copy()
```

```python
# 클러스터링이 진행되고나서 진행된 값을 cluster라는 컬럼에 넣음
group["cluster"] = model.predict(x_scaled)
```

```python
group = group.sort_values("cluster").reset_index(drop=True)
```

```python
group
```

```python
group["cluster"].value_counts()
```

### 클러스터링 결과
- 3개로 자치구를 나눴을 때 0번 : 11개, 1번 11개, 2번 3개 이렇게 클러스터링 결과가 나옴

## 각 클러스터들을 특징 파악
- 각 클러스터들의 평균을 구하고 비교 분석

```python
# cluster_means는 각각 클러스터의 평균
# all_means는 전체 평균
cluster_means = group.groupby("cluster")[select].mean(numeric_only=True)
all_means = group[select].mean().to_frame().T
```

```python
all_means
```

```python
cluster_means
```

```python
# 3개의 클러스터를 100분율로 계산
share = cluster_means.div(cluster_means.sum(axis=0), axis=1) * 100

share.T.plot(kind="bar", figsize=(12,6)) #bar형태로 그림

plt.title("클러스터별 지표 점유율 (%)")
plt.xlabel("지표")
plt.ylabel("점유율 (%)")
plt.legend(title="Cluster")
plt.tight_layout()

plt.show()
```

## 컬럼별 분석

## |대여-반납|

```python
metric = "|대여-반납|"

# 전체 평균(ALL) 스칼라 추출
overall = all_means.squeeze()[metric]

# 클러스터별 평균 Series
cm = cluster_means[metric]

# 간단한 요약(개수)
above = (cm > overall).sum()
below = (cm < overall).sum()
equal = (cm == overall).sum()

# 막대그래프
labels = ["ALL"] + [str(c) for c in cm.index]
values = [overall] + cm.tolist()

x = np.arange(len(labels))
plt.figure(figsize=(8,5))
plt.bar(x, values)
plt.xticks(x, labels)
plt.ylabel(metric)
plt.title(f"{metric} — 전체 평균 vs 클러스터 평균(원값)")

# 값 라벨 표시
for xi, v in zip(x, values):
    plt.text(xi, v, f"{v:,.0f}", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.show()
```

# 활성시간, 중간시간, 비활성시간 종합

## 활성시간 클러스터별 특징
- 0번 클러스터
  - 클러스터 중 회전율이 가장 높고, 시간대 최대-최소 편차가 크게 벌어지는 그룹
  - 피크 시간대에 수요가 급격히 치솟고 변동성이 높은 지역구들

- 1번 클러스터
  - 특정 지표가 두드러지지 않는 안정적인 그룹
  - 전반적으로 과열되지도 않는 기본적인 지역구들

- 2번 클러스터
  - 대부분의 지표가 월등히 높아 수요 불균형이 뚜렷하게 나타나는 그룹
  - 따릉이 불균형이 있어 지속적인 재배치가 필요한 지역구들

- 활성시간때는 2번 클러스터가 매우 불안정
  - 회전율(이용량)이 다른 클러스터에 비해 낮지만 |대여-반납|, 빈 대여소의 비율, 이동량이 매우 높음

- 활성시간때의 2번 클러스터에는 중구, 종로구, 강남구가 존재

## 중간시간 클러스터별 특징
- 0번 클러스터
  - 이동량, 회전율은 중간 다른 불균형 수치도 중간, 빈대여소 비율/시간 낮은 편
  - 구조적으로 공급부족이 크지는 않음, 안정적임

- 1번 클러스터
  - 빈 대여소의 비율과 빈 대여소의 시간이 가장 높음
  - 이동량 회전율은 낮음 상시 보충, 순회 거치대/재고 증설

- 2번 클러스터
  - 불균형, 이동량, 회전율이 최고 수준
  - 규모도 크고 회전이 빨라 실시간 재배치가 필요

- 중간시간때는 컬럼별로 특이값이 존재하는 클러스터가 2개 존재
  - 1번 클러스터 -> 이동량은 낮음, 빈대여소의 비율/시간이 높음
  - 2번 클러스터 -> 회전율, 이동량이 높음 |대여 - 반납|의 비율도 높음

## 비활성시간 클러스터별 특징
- 0번 클러스터
  - 이동량이 많고 회전율은 중상, 빈대여소 문제도 중간
  - 수요 규모는 꽤 되지만 구조적 불균형은 심하지 않은 지역구들

- 1번 클러스터
  - 빈 대여소 비율, 지속시간이 가장 높고 회전율 이동량은 낮은 공급부족형
  - 전반 지표가 과열은 아니지만 재고 고갈이 잦아 지속적인 재배치가 필요

- 2번 클러스터
  - 회전율, 최대-최소 편차 이동량이 모두 최고조로 불균형이 큼
  - 피크 시간대 급등락과 변동성이 매우 두드러짐

- 비활성시간 또한 중간시간때와 마찬가지로 컬럼별로 특이값이 존재하는 클러스터가 2개 존재
  - 1번 클러스터 -> 회전율은 낮고 빈대여소 비율/시간이 높음
  - 2번 클러스터 -> 회전율, 이동량, |대여-반납|이 최고조임

- 활성시간을 제외하고 2개의 불안정한 클러스터가 보임

- 이동량이 제일 크고 수요 불안정이 1개로 정확하게 나타난 활성시간의 2번 클러스터에는 중구, 종로구, 강남구가 존재
   - 종로구
     - 중간시간 : 1번클러스터
     - 비활성시간 : 1번클러스터
   - 중구
     - 중간시간 : 1번 클러스터
     - 비활성시간 : 1번 클러스터
   - 강남구
     - 중간시간 : 1번 클러스터
     - 비활성시간 : 1번 클러스터

## 종로구, 중구, 강남구의 공통점(특징)
- 활성시간때의 수요 불안정이 나타난 구들은 다른 구들에서 회전율은 낮고 빈대여소 비율/시간이 높은 특이점이 나옴
- 이용량이 높을때는 빈 대여소의 비율이 높고, |대여-반납| 차이, 이동량의 차이가 매우 높게 나타나지만 이용량이 적을때는 다른 클러스터들에 비해 전체적인 이용량이 적고 빈 대여소의 비율/시간만 높게 나타남
  - 활성시간은 출, 퇴근시간을 모두 포함하고 있음
    - |대여-반납|이 크다는 것은 대여 비율 또는 반납 비율이 높다는 것
    - 이동량이 높다는 것은 다른 자치구로 이동하거나 다른 자치구로부터 이동한 양이 많다는 것
    - 빈대여소 비율이 높다는 것은 수요에 비해서 자전거의 수가 적거나, 이 시간대의 자전거가 갑자기 많이 빠져나가거나 대여가 많이 됬다는 뜻
  - 비활성시간, 중간시간은 출, 퇴근시간을 제외함
    - 출,퇴근시간을 제외하고는 |대여-반납|이 한쪽으로 쏠리지 않음
    - 다른 구의 이동량이 많지 않음
    - 하지만 빈대여소의 비율/시간은 아직 높음, 수요에 비해서 자전거의 수가 적음


```python
df1 = pd.read_csv("./data/columns/merged_활성시간.csv")
df2 = pd.read_csv("./data/columns/merged_비활성시간.csv")
df3 = pd.read_csv("./data/columns/merged_중간시간.csv")
```

```python
df1 = df1[df1["자치구"].isin(["중구", "종로구", "강남구"])]
df2 = df2[df2["자치구"].isin(["중구", "종로구", "강남구"])]
df3 = df3[df3["자치구"].isin(["중구", "종로구", "강남구"])]
```

```python
df_merged = pd.concat([df1, df2, df3], axis = 0)
```

```python
df_merged = df_merged.sort_index()
```

```python
df_merged = df_merged.drop(["std"], axis = 1)
```

```python
df_merged
```

# |대여 - 반납|

```python
# 표만들기
cities = ["강남구", "종로구", "중구"]
torder = ["활성", "중간", "비활성"]

view = (df_merged
        .loc[df_merged["자치구"].isin(cities) & df_merged["시간_구분"].isin(torder),
             ["자치구","시간_구분","|대여-반납|"]]
        .copy())


view["시간_구분"] = pd.Categorical(view["시간_구분"], categories=torder, ordered=True)


table = (view
         .pivot_table(index="시간_구분", columns="자치구",
                      values="|대여-반납|", aggfunc="first")
         .reindex(torder))

print(table)
```

```python
ax = table[cities].plot(kind="bar", figsize=(8,4))
ax.set_title("시간_구분 × 자치구의 |대여-반납| (원시값)")
ax.set_xlabel("시간_구분"); ax.set_ylabel("|대여-반납|")
plt.legend(title="자치구")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 빈대여소의비율

```python
import pandas as pd

cities = ["강남구", "종로구", "중구"]
torder = ["활성", "중간", "비활성"]

view = (df_merged.loc[
            df_merged["자치구"].isin(cities) & df_merged["시간_구분"].isin(torder),
            ["자치구","시간_구분","빈대여소의비율"]
        ].copy())

# 시간 순서 고정
view["시간_구분"] = pd.Categorical(view["시간_구분"], categories=torder, ordered=True)

# 같은 (자치구, 시간_구분) 조합이 1행이라고 가정 → 'first'
table_ratio = (view.pivot_table(index="시간_구분", columns="자치구",
                                values="빈대여소의비율", aggfunc="first")
                    .reindex(torder))

print(table_ratio)
```

```python
import matplotlib.pyplot as plt

ax = table_ratio.plot(kind="bar", figsize=(8,4))
ax.set_title("시간_구분 × 자치구의 빈대여소의비율")
ax.set_xlabel("시간_구분"); ax.set_ylabel("빈대여소의비율")
plt.legend(title="자치구")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# 이동량

```python
import pandas as pd
import matplotlib.pyplot as plt

cities = ["강남구", "종로구", "중구"]
torder = ["활성", "중간", "비활성"]

ax = (df_merged
      .query('자치구 in @cities and 시간_구분 in @torder')
      .assign(시간_구분=lambda d: pd.Categorical(d["시간_구분"], categories=torder, ordered=True))
      .pivot_table(index="시간_구분", columns="자치구",
                   values="이동량", aggfunc="first")
      .reindex(torder)
      .plot(kind="bar", figsize=(8,4))
)

ax.set_title("시간_구분 × 자치구의 이동량")
ax.set_xlabel("시간_구분"); ax.set_ylabel("이동량")
ax.legend(title="자치구")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

cities = ["강남구", "종로구", "중구"]
torder = ["활성", "중간", "비활성"]

ax = (df_merged
      .query('자치구 in @cities and 시간_구분 in @torder')
      .assign(시간_구분=lambda d: pd.Categorical(d["시간_구분"], categories=torder, ordered=True))
      .pivot_table(index="시간_구분", columns="자치구",
                   values="회전율", aggfunc="mean")   # 여러 행이면 평균이 더 자연스러움(필요시 median/sum)
      .reindex(torder)
      .plot(kind="bar", figsize=(8,4))
)

ax.set_title("시간_구분 × 자치구의 회전율")
ax.set_xlabel("시간_구분"); ax.set_ylabel("회전율")
ax.legend(title="자치구")
plt.xticks(rotation=0)

# 회전율이 0~1 스케일이면 퍼센트로 표시
col = "회전율"
if df_merged[col].dropna().between(0, 1).all():
    ax.yaxis.set_major_formatter(PercentFormatter(1.0))

plt.tight_layout()
plt.show()
```
