```python
import pandas as pd


df_location = pd.read_csv('6월 데이터.csv')

df_target = pd.read_csv('data_2406 (1).csv')

df_location.rename(columns={'대여소 번호': '대여소번호', '대여소 위치(구)': '자치구'}, inplace=True)

df_location['대여소번호'] = pd.to_numeric(df_location['대여소번호'], errors='coerce')
df_location.dropna(subset=['대여소번호'], inplace=True)
df_location['대여소번호'] = df_location['대여소번호'].astype(int)

df_location = df_location[['대여소번호', '자치구']].drop_duplicates()

df_target = df_target.drop(columns=['자치구'])

merged_df = pd.merge(df_target, df_location, on='대여소번호', how='left')

print("병합 결과 확인:")
print(merged_df.head())

output_filename = 'merged_data_final.csv'
merged_df.to_csv(output_filename, index=False, encoding='utf-8-sig')

print(f"\n작업 완료! '{output_filename}' 파일이 저장되었습니다.")
```
