# QA

Extractive QA (추출형) : 주어진 문맥에서 직접 추출되는 방식의 작업

- 문서 내에 명시된 정보를 기반으로 정답을 찾아내는 유형
- 질문의 답이 항상 문맥 내에 존재하며, 모델이 해당 답을 위치별로 찾아내는 과정

```python
dataset = datasets.load_dataset('squad', split = "train", streaming=True)
```

일반적인 데이터 구조
- context : 질문에 대한 정답이 포함된 문서 또는 텍스트
- question : 사용자가 묻는 질문
- answers : 주어진 문맥에서 추출된 정답(단어 또는 문장)
- answer_start : 문맥에서 정답이 시작되는 위치

```python
temp = None
for i, example in enumerate(dataset):
    temp = example
    break

print(temp)
```

Generative QA (생성형) : 주어진 문맥에 근거하여 정답을 생성하는 task

```python
dataset = datasets.load_dataset('natural_questions', split = "train", streaming=True)
```
