# **[멋쟁이사자처럼_AI_NLP] 02. 인공지능 체험하기**

## 1. 생성형 AI

- 생성형 AI는 텍스트, 이미지, 비디오 등 다양한 형태의 콘텐츠를 생성하는 인공지능 기술
- 주로 정보 제공, 창작, 요약, 번역, 개인화된 추천 등의 기능을 수행

### ChatGPT (OpenAI)

- 자연어 처리 기반의 대화형 AI
- 질문에 대한 답변, 정보 제공, 창작물 생성(스토리, 시나리오 등) 등을 수행하며 사용자의 요청에 맞춰 유연하게 대화를 이어간다.

### DALL-E

- 원하는 이미지를 텍스트로 입력하면 이를 기반으로 이미지를 생성하는 AI 서비스

## 2. 할루시네이션

- AI 모델이 실제 사실과 다르거나 존재하지 않는 정보를 그럴듯하게 생성하는 현상
- 모델이 학습한 데이터의 일부를 바탕으로 현실과 일치하지 않는 답변을 만드는 것

### 할루시네이션이 발생하는 원인

- **훈련 데이터의 한계**: 모델은 대규모 데이터에서 패턴을 학습하지만, 이 데이터가 불완전하거나 부정확할 경우 잘못된 정보가 생성될 수 있다.
- **추론 과정의 복잡성**: 모델이 문맥이나 질문을 잘못 이해할 때, 잘못된 결론에 도달할 수 있다.
- **임의성(Randomness)**: 생성형 AI는 때때로 창의성을 발휘하도록 설계되었으며, 이 과정에서 사실과 다른 내용을 생성할 수 있다.

### 생성형 AI 사용을 위한 조언

- 창의성이 필요한 작업과 정밀성이 필요한 작업을 구분
- 생성된 내용에 대하여 반드시 검증
- 명확하고 구체적인 프롬프트 사용
- 구체적인 피드백 적용

## 3. 프롬프트 엔지니어링

- 생성형 AI는 텍스트를 입력받아 동작을 수행하고 이러한 텍스트
- 프롬프트 엔지니어링은 AI 모델에게 입력하는 프롬프트를 설계하여 최적의 결과를 도출하는 기술

### 프롬프트 작성 시 고려해야할 것

- **명확성 (Clarity)**
	- 요청을 구체적으로 작성하여 원하는 정보를 명확히 전달한다.
	- (나쁜 예) "관광지를 추천해줘."
	- (좋은 예) "8월에 강원도에서 부모님과 방문할 만한 곳을 추천해줘."
- **맥락 (Context)**
	- 필요한 경우 추가적인 배경 정보를 포함하여 모델이 더 잘 이해할 수 있도록 한다.
	- (예) "부모님이 걷는 것을 힘들어하시니, 차량으로 이동하기 편한 곳 위주로"

- **결과물의 형식 (Format)**
	- 원하는 답변의 형식을 구체적으로 지정한다.
	- (예) "추천하는 10곳의 관광지를 '장소명', '특징'을 포함한 표(Table) 형식으로 정리해줘."

- **역할 설정 (Role-Playing)**
	- 특정한 상황이나 역할을 부여하여 모델이 그에 맞는 답변을 하도록 유도한다.
	- (예) "당신은 30년 경력의 여행 가이드입니다. 여행 경비를 고려해서..."

- **피드백 (Feedback)**
	- 결과물을 확인한 뒤 피드백을 통해 프롬프트를 수정하고 개선하여 더 나은 결과를 얻는다.
    
### API 사용시 작성하는 프롬프트 예시

```
{
  "model": "<모델명>",
  "messages": [
    {
      "role": "system",
      "content": "<모델이 맡을 역할>"
    },
    {
      "role": "user",
      "content": "<이전 질문>"
    },
    {
      "role": "assistant",
      "content": "<이전 답변>"
    },
    {
      "role": "user",
      "content": "<현재 질문>"
    }
  ],
  "temperature": 0.7,
  "max_tokens": 4096,
  "frequency_penalty": 0.1,
  "presence_penalty": 0.6,
}
```

- `messages`: 각 메시지는 **역할(role)**과 **내용(content)**이 있다.
	- `system`: 모델이 수행할 기본 역할(예: "너는 여행 비용 관리자야.")을 지정.
	- `user`: 사용자의 질문이나 요청 사항을 전달.
	- `assistant`: 이전 대화의 맥락을 유지하기 위해 모델의 이전 답변을 포함.
- `temperature`: 모델의 창의성(무작위성)을 조절. (0 ~ 1, 높을수록 창의적)
- `max_tokens`: 생성할 답변의 최대 길이를 제한.
- `frequency_penalty`: 동일한 단어의 반복 사용을 제어. (높을수록 반복 감소)
- `presence_penalty`: 새로운 주제나 정보를 제공하도록 제어. (높을수록 새로운 주제)
